```python
enderecos = pd.read_csv("/content/drive/MyDrive/data/Imersao_dados_4/enderecos.csv")
enderecos.head()
```

Esta base de dados endereços tem valores que são de fora de SP pra agilizar buscas futuras vamos filtar os dados apenas para SP

```python
enderecos_sp = enderecos.query("sigla_uf == 'SP' ")
enderecos_sp.head()
```

Agora não podemos trabalhar com a coluna logradouro completo (logr_compl) pois ela possui alguns problemas como por exemplo: Rua João Antônio Vasconcelos - até 627/628. Isto irá gerar problemas futuros, a solução é trbalhar com o tipo logradouro e logradouro nome, respectivamente "tipo_logr" e	"logr_nome". Para tal vamos criar uma nova variável neste dataframe, chamada "rua" que combina estas duas variáveis	 

```python
enderecos_sp["rua"] = enderecos_sp["tipo_logr"] + " " + enderecos_sp["logr_nome"] # combina os valores nas mesmas linhas das duas variáveis com um espaço entre eles
enderecos_sp["rua"] = enderecos_sp["rua"].str.lower().str.strip() # transforma os valores presentes na linha em letras minúsculas e depois remove os espaços extras que possam existir (para evitar problemas no futuro)
enderecos_sp.head()
```

Vamos fazer algo parecido para os dados de vendas (data_sell), porém para este dataframe temos o número da casa junto ao logradouro. Teremos então que remover o número do logradouro, para ficar apenas com o nome das ruas

```python
data_sell.head(3)
```

```python
data_sell["Rua"].str.extract(r'(^[\w ]+)')[:10] # faz uma extração (rejects) de tudo o que for letra mais um espaço e mostra as 10 primeiras linhas
```

```python
data_sell["apenas_rua"] = data_sell["Rua"].str.extract(r'(^[\w ]+)')
data_sell["apenas_rua"] = data_sell["apenas_rua"].str.lower().str.strip()
data_sell.head()
```

Agora vamos vincular as duas bases a de vendas (data_sell) com a de endereços (enderecos_sp), deste ultimo vamos pegar apenas as colunas CEP, Latitude e Longitude. Para isto a função "merge" do pandas será usada e criaremos uma nova base de dados chamada data_geo 

```python
data_geo = pd.merge(left = data_sell, right = enderecos_sp[["rua","cep","latitude", "longitude"]], how= "left", left_on="apenas_rua", right_on="rua")
#os parametros left_on e right_on se referem a coluna que será o link entre as 2 bases de dados, ou seja os dados que são iguais entre elas.
data_geo
```

Observando o Dataframe notou-se que varias linhas foram criadas com informaçoes muito similares, pode ser devido a terem o nome de rua igual, porém latitudes e longitudes diferentes e etc. Ou seja está duplicando várias linhas, ou seja precisamos remover estas duplicatas. Deste modo vamos manter uma única linha para cada dado que estiver na tabela de vendas (data_sell)

```python
data_geo = pd.merge(left = data_sell, right = enderecos_sp[["rua","cep","latitude", "longitude"]], how= "left", left_on="apenas_rua", right_on="rua").drop_duplicates(subset=data_sell.columns).query("cep > 0") 
data_geo
```

Nota-se que o número de registros foi reduzido drasticamente, o dataframe com duplicatas tinha 90280 linhas e o atual "limpo" possui apenas 8520 linhas

Algumas métricas que visam resumir como o modelo performa serão calculadas, são elas o R² e o erro médio absoluto (mean absolut error)
O primeiro passo será importar estas funçoes das bibliotecas

```python
from sklearn.metrics import mean_absolute_error, r2_score

mean_absolute_error(Y_test, Y_predict_test) #vamos passar o valor real e o valor predito
```

Ele retornou um número bastante alto, a interpretação deste valor é: Quão maior este número maior é o erro do modelo. Ou seja ele está mais distante dos valores preditos.

Só para ter um comparativo vamos pegar os valores de X e Y de treino


```python
mean_absolute_error(Y_train, Y_predict_train)
```

Analisando estes valores pode se observar que o modelo apresenta mais erros utilizando os dados de treino que os dados preditos. Este fenômeno é chamado de underfitting que ele não esta bom nem para os valores em que foi treinado. O que indica que apenas um valor para treinar o modelo não é suficiente para resolvar o problema, que é predizer os valores dos imóveis.

Teremos então que retreinar o modelo adicionando mais varíaveis


```python
X = dados_vendas_censo[['Metragem', 'Quartos', 'Banheiros', 'Vagas','latitude', 'longitude', 'Situacao_setor', 
                        'V001','V002', 'V003', 'V004', 'V005', 'V006','V007','V008', 'V009', 'V010', 'V011', 'V012',]]
                        #As variáveis valor do anuncio e valor por m2 ficaram de fora, pois o valor por m2 acabava vazando o valor da resposta (Data leak)

Y = dados_vendas_censo["Valor_anuncio"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 54)
lr.fit(X_train, Y_train)
Y_predict_test = lr.predict(X_test)
Y_predict_train = lr.predict(X_train)
mae_test = mean_absolute_error(Y_test, Y_predict_test) #calculando o erro médio absoluto
r2_test = r2_score(Y_test, Y_predict_test) #calculando o valor do R²
mae_train = mean_absolute_error(Y_train, Y_predict_train)
r2_train = r2_score(Y_train, Y_predict_train)
```

Verificando o valor do erro médio absoluto e do R² de teste

```python
mae_test
```

```python
r2_test
```

Comparando agora ao valor do erro médio absoluto e do R² de **predição**

```python
mae_train
```

```python
r2_train
```

Avaliando os resultados percebe-se que pelo valor de R² e do MAE (Mean Absolute error) o valor do modelo treinado é um pouco melhor que os valores usados como teste. Em suma o modelo performa melhor.
